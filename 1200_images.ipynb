```python
# 2012
from PIL import Image
import os

# 提高像素限制
Image.MAX_IMAGE_PIXELS = None

# 设置源图片的路径
src_path = "/Users/huanjingheng/Downloads/STATS-402/Building_change_detection_dataset/Building change detection dataset_add/1. The two-period image data/2012/whole_image/train/image/2012_train.tif"

# 设置存储分割后的图片的路径
dst_path = "/Users/huanjingheng/Downloads/STATS-402/new_splitted_images/2012"

# 读取图片
src_img = Image.open(src_path)

# 图片的大小
src_width, src_height = src_img.size

# 切割的大小
dst_size = (500, 500)

# 计数器
count = 0

for i in range(0, src_width - dst_size[0] + 1, dst_size[0]):
    for j in range(0, src_height - dst_size[1] + 1, dst_size[1]):
        box = (i, j, i+dst_size[0], j+dst_size[1])
        dst_img = src_img.crop(box)
        
        # 创建文件名
        filename = f"2012_{i}_{j}.jpg"
        
        # 保存图片
        dst_img.save(os.path.join(dst_path, filename))
        
        count += 1
        
        # 停止条件
        if count >= 1200:
            break
    if count >= 1200:
        break

print(f"Total {count} images are created.")
```
